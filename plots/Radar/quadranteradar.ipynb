```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

image_files = ['./rodada01/Everton.png', './rodada01/jeffinho.png', './rodada01/Leandrinho.png', './rodada01/Lele.png']
# Criar diretório se não existir
output_directory = './rodada01/'
os.makedirs(output_directory, exist_ok=True)

# Criar o plot
fig, axs = plt.subplots(2, 2, figsize=(12, 12), facecolor= 'black')  # Ajuste o tamanho para 16x16
for i in range(2):
    for j in range(2):
        img_index = i * 2 + j
        img = mpimg.imread(image_files[img_index])
        axs[i, j].imshow(img)
        axs[i, j].axis('off')

# Ajustar o espaçamento entre os subplots
plt.subplots_adjust(left=0, right=1, bottom=0, top=1, wspace=0, hspace=0)

# Salvar o plot no diretório especificado
plt.savefig(os.path.join(output_directory, 'combined_plot.png'), dpi=400)  # Ajuste o valor de dpi conforme necessário

# Exibir o plot
plt.show()
```
