```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the first CSV file and sort by 'Ground_duels_won'
df1 = pd.read_csv("./CSV/rodada1/fluminensexvre_01/flu.csv").sort_values(by='Ground_duels_won', ascending=False)
languages1 = df1['Player'].tolist()
ground_duels1 = df1['Ground_duels_won'].tolist()
aerial_duels1 = df1['Aerial_duels_won'].tolist()

# Read the second CSV file and sort by 'Ground_duels_won'
df2 = pd.read_csv("./CSV/rodada1/fluminensexvre_01/vre.csv").sort_values(by='Ground_duels_won', ascending=False)
languages2 = df2['Player'].tolist()
ground_duels2 = df2['Ground_duels_won'].tolist()
aerial_duels2 = df2['Aerial_duels_won'].tolist()

# Combine data from both scatter plots
languages = languages1 + [''] + languages2
ground_duels = ground_duels1 + [np.nan] + ground_duels2
aerial_duels = aerial_duels1 + [np.nan] + aerial_duels2

# Plotting the horizontal lollipop chart with black background
fig, ax = plt.subplots(figsize=(12, 8))  # Increase the figure size
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

# Plotting ground statistics
ax.hlines(y=languages, xmin=0, xmax=ground_duels, color='white', alpha=0.7, linewidth=2)
ax.scatter(x=ground_duels, y=languages, s=75, color='white', alpha=0.7, label='Ground Duels')

# Plotting aerial statistics
ax.hlines(y=languages, xmin=0, xmax=aerial_duels, color='green', alpha=0.7, linewidth=2)
ax.scatter(x=aerial_duels, y=languages, s=75, color='green', alpha=0.7, label='Aerial Duels')

# Vertical lines to mark the boundaries
ax.vlines(x=max(ground_duels1), ymin=min(languages1), ymax=max(languages1 + languages2), color='white', linestyle='--')

# Labeling
ax.set_yticks(languages)
ax.set_yticklabels(languages, color='white')
ax.tick_params(axis='x', colors='white')  # Set x-axis tick color to white
ax.set_xticks(np.arange(0, max(ground_duels)+1))  # Explicitly set x-axis ticks from 0 to 20
ax.set_xlim(0, max(max(ground_duels1), max(aerial_duels1)) + 5)  # Adjust x-axis limit
max_value = max(max(ground_duels), max(aerial_duels))
ax.set_xlim(0, max_value + 5) 
# Adding labels and title
plt.ylabel('Players', color='white')
plt.xlabel('Duel Values', color='white')
plt.title('Fluminense x Volta Redonda | Round 1 - Carioca Championship', color='white')
plt.suptitle('Ground duel and Aerial duels won by player', color='white')

# Adding legend with white text
legend = ax.legend(facecolor='black', edgecolor='white')
for text in legend.get_texts():
    text.set_color('white')

# Adding grid to the x-axis
ax.grid(axis='x', linestyle='--', alpha=0.7, color='white')

# Save the plot to the specified directory
plt.savefig('./plots/groundxaerial/rodada01/fluminensexvre_01.png', facecolor=fig.get_facecolor(), bbox_inches='tight')  # Add bbox_inches='tight'

# Displaying the lollipop chart
plt.show()
```
